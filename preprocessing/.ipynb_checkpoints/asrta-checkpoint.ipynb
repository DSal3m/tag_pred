```python
import re
import pandas as pd

# Функции для предобработки кода на C++
def remove_cpp_comments_and_metadata(code):
    code = re.sub(r'https?://\S+', '', code)
    code = re.sub(r'//.*', '', code)
    code = re.sub(r'/\*.*?\*/', '', code, flags=re.DOTALL)
    code = re.sub(r'// Problem:.*?// Contest:.*?\n', '', code)
    code = re.sub(r'#include.*?\n', '', code)
    code = re.sub(r'#define.*?\n', '', code)
    code = re.sub(r'#ifndef.*?#endif', '', code, flags=re.DOTALL)
    code = re.sub(r'#ifndef ONLINE_JUDGE.*?#endif', '', code, flags=re.DOTALL)
    code = re.sub(r'using namespace std;', '', code)
    code = re.sub(r'\s+', ' ', code)
    code = re.sub(r'#\w+.*', '', code)
    return code

def normalize_cpp_indents(code):
    code = code.replace('\t', '    ')
    code = re.sub(r'^\s+|\s+$', '', code, flags=re.MULTILINE)
    return code

def replace_strings(code):
    # Replace integers and floating-point numbers with <NUM>
    code = re.sub(r'".*?"', '<STR>', code)
    code = re.sub(r"'.*?'", '<CHAR>', code)
    return code

def replace_numericals(code):
    # Replace integers and floating-point numbers with <NUM>
    code = re.sub(r'\b\d+\b', '<NUM>', code)  # Integer literals
    code = re.sub(r'\b\d+\.\d+\b', '<NUM>', code)  # Floating-point literals
    return code

def preprocess_cpp_code(code):
    code = remove_cpp_comments_and_metadata(code)
    code = normalize_cpp_indents(code)
    code = replace_numericals(code)
    code = replace_strings(code)
    return code

# Функции для предобработки кода на Java
def remove_java_comments_and_metadata(code):
    code = re.sub(r'https?://\S+', '', code)
    code = re.sub(r'//.*', '', code)
    code = re.sub(r'/\*.*?\*/', '', code, flags=re.DOTALL)
    code = re.sub(r'package.*?;', '', code)
    code = re.sub(r'import.*?;', '', code)
    code = re.sub(r'// Problem:.*?// Contest:.*?\n', '', code)
    code = re.sub(r'System.set(In|Out)\(.*?ONLINE_JUDGE.*?\);', '', code)
    return code

def normalize_java_indents(code):
    code = code.replace('\t', '    ')
    code = re.sub(r'^\s+|\s+$', '', code, flags=re.MULTILINE)
    return code

def preprocess_java_code(code):
    code = remove_java_comments_and_metadata(code)
    code = normalize_java_indents(code)
    code = replace_numericals(code)
    code = replace_strings(code)
    return code

cpp_code_path = "C:/Users/Alex/Downloads/cpp_solutions.csv"
java_code_path = "C:/Users/Alex/Downloads/java_solutions.csv"

# Загрузка и предобработка кода C++
df_cpp = pd.read_csv(cpp_code_path)
df_cpp['Code'] = df_cpp['Code'].apply(preprocess_cpp_code)

# Загрузка и предобработка кода Java
df_java = pd.read_csv(java_code_path)
df_java['Code'] = df_java['Code'].apply(preprocess_java_code)

cpp_cleaned_code_path = "C:/Users/Alex/Downloads/cpp_solutions_cleaned.csv"
java_cleaned_code_path = "C:/Users/Alex/Downloads/java_solutions_cleaned.csv"

df_cpp.to_csv(cpp_cleaned_code_path, index=False)
df_java.to_csv(java_cleaned_code_path, index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt


df_cpp = pd.read_csv("C:/Users/Alex/Downloads/cpp_solutions_cleaned.csv")
df_java = pd.read_csv("C:/Users/Alex/Downloads/java_solutions_cleaned.csv")


cpp_code_lengths = df_cpp['Code'].str.len()
java_code_lengths = df_java['Code'].str.len()


plt.figure(figsize=(10, 6))
plt.hist(cpp_code_lengths, bins=200, range=(0, 20000), alpha=0.7, label='C++ Code Lengths')
plt.hist(java_code_lengths, bins=200, range=(0, 20000), alpha=0.7, label='Java Code Lengths')
plt.title('Распределение длин кодов C++ и Java до 20k символов')
plt.xlabel('Длина кода (в символах)')
plt.ylabel('Количество фрагментов кода')
plt.legend()
plt.show()
```

```python
# Определение количества строк в каждом фрагменте кода
cpp_lines = df_cpp['Code'].str.count('\n') + 1
java_lines = df_java['Code'].str.count('\n') + 1

# Построение гистограммы распределения количества строк
plt.figure(figsize=(10, 6))
plt.hist(cpp_lines, bins=50, range=(0, 500), alpha=0.7, label='C++ Code Lines')
plt.hist(java_lines, bins=50, range=(0, 500), alpha=0.7, label='Java Code Lines')
plt.title('Распределение количества строк кода C++ и Java до 20k строк')
plt.xlabel('Количество строк')
plt.ylabel('Частота')
plt.legend()
plt.show()
```
